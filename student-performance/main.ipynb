```python
import kagglehub
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Download latest version
path = kagglehub.dataset_download("joshuanaude/effects-of-alcohol-on-student-performance", path="Stats survey.csv")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv(path)

df['Were you on scholarship/bursary in 2023?'] = df['Were you on scholarship/bursary in 2023?'].map({
    'No': 0,
    'Yes (NSFAS, etc...)': 1
})

df['Additional amount of studying (in hrs) per week'] = df['Additional amount of studying (in hrs) per week'].map({
    '0': 0,
    '1-3': 1,
    '3-5': 2,
    '5-8': 3,
    '8+': 4
})

df['On a night out, how many alcoholic drinks do you consume?'] = df['On a night out, how many alcoholic drinks do you consume?'].map({
    '0': 0,
    '1-3': 1,
    '3-5': 2,
    '5-8': 3,
    '8+': 4
})

df['How many classes do you miss per week due to alcohol reasons, (i.e: being hungover or too tired?)'] = df['How many classes do you miss per week due to alcohol reasons, (i.e: being hungover or too tired?)'].map({
    '0': 0,
    '1': 1,
    '2': 2,
    '3': 3,
    '4+': 4
})

df['How many modules have you failed thus far into your studies?'] = df['How many modules have you failed thus far into your studies?'].map({
    '0': 0,
    '1': 1,
    '2': 2,
    '3': 3,
    '4+': 4
})

df['Are you currently in a romantic relationship?'] = df['Are you currently in a romantic relationship?'].map({
    'Yes': 0,
    'No': 1
})

df['Do your parents approve alcohol consumption?'] = df['Do your parents approve alcohol consumption?'].map({
    'Yes': 0,
    'No': 1
})

df['How strong is your relationship with your parent/s?'] = df['How strong is your relationship with your parent/s?'].map({
    'Very close': 0,
    'Close': 1,
    'Fair': 2,
    'Distant': 3
})

df['Your Accommodation Status Last Year (2023)'] = df['Your Accommodation Status Last Year (2023)'].map({
    'Non-private accommodation ie. Res': 1,
    'Private accommodation/ stay with family/friends': 0,
})

df['How often do you go out partying/socialising during the week? '] = df['How often do you go out partying/socialising during the week? '].map({
    'Only weekends': 0,
    '0': 0,
    '1': 1,
    '2': 2,
    '3': 3,
    '4+': 4,
})

df['Monthly Allowance in 2023'] = df['Monthly Allowance in 2023'].map({
    'R 4001- R 5000': 0,
    'R 5001 - R 6000': 1,
    'R 6001 - R 7000': 2,
    'R 7001 - R 8000': 3,
    'R 8000+': 4,
})

df['What year were you in last year (2023) ?'] = df['What year were you in last year (2023) ?'].map({
    '1st Year': 0,
    '2nd Year': 1,
    '3rd Year': 2,
    '4th Year': 3,
    'Postgraduate': 4,
})


df['Your Sex?'] = df['Your Sex?'].map({
    'Female': 0,
    'Male': 1
})
```
